```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

df['remoto'].value_counts()
```

```python
import numpy as np
```

```python
# Cria o Data Frame
df_salarios = pd.DataFrame({
    'nome': ['Anna', 'Gabriel', 'Pedro', 'Leo', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# Preenche os valores nulos com a media dos salarios arredondada
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Preenche os valores nulos com a mediana dos salarios
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

# Printa o Data Frame inteiro
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Anna', 'Pedro', 'Gabriel', 'Leo', 'Val'],
    'cidade': ['Itajuba', np.nan, 'Três Corações', np.nan, 'Varginha']
})

# .fillna()
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

```python
data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
data_scientist
```

```python
data_scientist['residencia'].head().reset_index()
```

```python
import pycountry

def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None
```

```python
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
data_scientist

data_scientist_salario = data_scientist.groupby('residencia_iso3')['usd'].mean().round(2).sort_values(ascending=False).reset_index()
data_scientist_salario.head()
```

```python
fig = px.bar(data_scientist_salario,
            x='residencia_iso3',
            y='usd',
            title ='Media Salarial de Data Scientist por Paises',
            labels={'residencia_iso3': 'Países', 'usd': 'Media salarial anual (USD)'},
            color='usd',
            template='plotly_white'
            )
fig.show()
```

```python
fig = px.choropleth(data_scientist_salario,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='PuBu',
                    title='Salário Médio de Cientista de Dados por país',
                    labels={'usd': 'Salário Médio (USD)', 'residencia_iso3': 'Países'})
fig.show()
```
